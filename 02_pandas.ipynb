```python
df_students_info = pd.read_csv('Data/Students_info_0.csv')
df_students_marks = pd.read_csv('Data/Students_marks_0.csv')

for i in range(1, 10):
  students_info_i = pd.read_csv('Data/Students_info_{}.csv'.format(i))
  df_students_info = pd.concat([df_students_info, students_info_i])              # concat the next file with students info with the previous one

  students_marks_i = pd.read_csv('Data/Students_marks_{}.csv'.format(i))
  df_students_marks = pd.concat([df_students_marks, students_marks_i])           # concat the next file with students marks with the previous one

df = pd.merge(df_students_info, df_students_marks, on='index', how='left')       # merge generated file_info with file_marks
df.sample(n=5)
```

#### 10.
1) Какая доля студентов сдала экзамен по математике (passmark >= 50)

2) Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?

3) Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

```python
# 1) percentage of students who passed the math exam:
percentage = len(df[df['math score'] >= passmark]) / len(df)

# 2) percentage of students who completed preparation course and passed the math exam:
prep_completed = len(df.loc[df['test preparation'] == 'completed'])
pass_with_prep = len(df.loc[(df["test preparation"] == 'completed')\
                          & (df["math score"] >= passmark)])

percentage_of_pass = pass_with_prep / prep_completed

# 3) percentage of female students who didn't have a preparation course and failed the math exam:
prep_none_female = len(df.loc[(df["test preparation"] == 'none')\
                            & (df["gender"] == 'female')])

prep_none_female_failed = len(df.loc[(df["test preparation"] == 'none')\
                                   & (df["gender"] == 'female')\
                                   & (df['math score'] < passmark)])

percentage_of_fail_female = prep_none_female_failed / prep_none_female

print('1)', percentage)
print('2)', percentage_of_pass)
print('3)', percentage_of_fail_female)
```

```python
%%time
ethnicity = sorted(set(df['race/ethnicity']))                                    # all the unique values of column 'race/ethnicity'

for ethnicity_i in ethnicity:                                                    # iterating by every ethnicity group

  by_ethnicity_data = df[df['race/ethnicity'] == ethnicity_i].reset_index(drop=True)  # creating dataframe for ethnicity group
  total_reading_score = 0

  for row in range(len(by_ethnicity_data)):                                      # summing up all the values of reading score in the rows
    total_reading_score += by_ethnicity_data.at[row, 'reading score']

  print(ethnicity_i, ' mean reading score: ', total_reading_score / len(by_ethnicity_data))
  # takes longer
```

```python
%%time
education = sorted(set(df['education']))                                         # all the unique values of column 'education'

for education_level in education:                                                # iterating by every education_level

  by_education_data = df[df['education'] == education_level].reset_index(drop=True) # creating dataframe for education_level
  writing_scores = []

  for row in range(len(by_education_data)):
    writing_scores.append(by_education_data.at[row, 'writing score'])            # adding the value of writing score in every row to a list

  print(education_level, '\tmin writing score:\t', min(writing_scores))          # minimum value in the list
  # takes longer
```

#### 15. Переведем баллы в оценки. Создадим вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным критериям:
- больше 90 = A
- 80-90 = B
- 70-80 = C
- 60-70 = D
- 50-60 = E
- меньше 50 = F (Fail)

```python
def GetGrade(average_mark):
    if average_mark > 90:
      return 'A'
    elif 80 < average_mark <= 90:
      return 'B'
    elif 70 < average_mark <= 80:
      return 'C'
    elif 60 < average_mark <= 70:
      return 'D'
    elif 50 < average_mark <= 60:
      return 'E'
    return 'F'

df['Grade'] = df.apply(lambda score: (score['math score'] + score['reading score'] + score['writing score']) / 3, axis=1).apply(GetGrade)
print(df['Grade'].value_counts()) # количество студентов, получивших каждую из оценок
df.head()
```
